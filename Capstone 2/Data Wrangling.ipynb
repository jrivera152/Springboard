```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#importing main data frame
df = pd.read_csv('diabetic_data.csv')
df.head()
```

```python
#importing data frame which maps IDs to a description 
#there are three sets of IDs in this df (admission_type, discharge_disposition, and admission_source) 
#that will need to be split up. this is done in the next few lines.
mapping_df = pd.read_csv('IDs_mapping.csv')
mapping_df.head(12)
```

```python
#creating df for admission_type mapping
admission_type = mapping_df.iloc[0:8]
admission_type.set_index('admission_type_id', drop=True, inplace=True)
admission_type = admission_type.transpose()
admission_type
```

```python
#turning df into a dictionary
admission_type_dict = admission_type.to_dict('list')
admission_type_dict
```

```python
#creating df for discharge_disposition mapping
discharge_disposition = mapping_df.iloc[10:40]
discharge_disposition.columns = ['discharge_disposition_id', 'description']
discharge_disposition.set_index('discharge_disposition_id', inplace=True, drop=True)
discharge_disposition = discharge_disposition.transpose()
discharge_disposition
```

```python
#turning df into df
discharge_disposition_dict = discharge_disposition.to_dict('list')
discharge_disposition_dict
```

```python
#creating df for admission_source mapping
admission_source = mapping_df.iloc[42:69]
admission_source.columns = ['admission_source_id', 'description']
admission_source.set_index('admission_source_id', inplace=True, drop=True)
admission_source = admission_source.transpose()
admission_source
```

```python
#turning df into df
admission_source_dict = admission_source.to_dict('list')
admission_source_dict
```

```python
df['admission_type_id'] =  df['admission_type_id'].astype(str).replace(admission_type_dict)
df['discharge_disposition_id'] =  df['discharge_disposition_id'].astype(str).replace(discharge_disposition_dict)  
df['admission_source_id'] =  df['admission_source_id'].astype(str).replace(admission_source_dict)  
```

```python
#printing the unique values for each column
col_list = ['race', 'gender', 'age', 'weight',
       'admission_type_id', 'discharge_disposition_id', 'admission_source_id',
       'time_in_hospital', 'payer_code', 'medical_specialty',
       'num_lab_procedures', 'num_procedures', 'num_medications',
       'number_outpatient', 'number_emergency', 'number_inpatient', 
       'number_diagnoses', 'max_glu_serum', 'A1Cresult',
       'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
       'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'examide', 'citoglipton', 'insulin',
       'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone', 'change', 'diabetesMed', 'readmitted']

print('Unique values in each column\n')
for col_name in col_list:
    unique_values = []
    
    for i in range(0, len(df[col_name].unique())):
        unique_values.append(df[col_name].unique()[i])
    
    print(col_name + ":")
    print(unique_values)
    print()
```

```python
#checking how many duplicate rows there are
df.duplicated().sum()
```

```python
#finding the percentage of missing values in each column
#for this, a missing value is considered a '?' as the value
for column in df.columns:
    percent_missing = np.sum(df[column]=='?')/df.shape[0]*100
    print(column + ': ' + str(percent_missing))
```

```python
#dropping weight, payer code, and medical specialty columns as they have significant number of missing values
df.drop(labels=['weight','payer_code','medical_specialty'], inplace=True, axis=1)
df.head()
```

```python
#dropping examide and citoglipton columns as they only have one unique value 'No'
df.drop(labels=['examide','citoglipton'], inplace=True, axis=1)
df.head()
```

```python
#replacing the ? with Other
df['race'].replace('?','Other', inplace=True)
df['race'].value_counts()
```

```python
#replacing the not available and not mapped values in the columns with the mode from the columns
df['admission_type_id'] =  df['admission_type_id'].replace({'Not Available':'Emergency','Not Mapped':'Emergency'})
df['discharge_disposition_id'] =  df['discharge_disposition_id'].replace({'Not Mapped':'Discharged to home'}) 
df['admission_source_id'] =  df['admission_source_id'].replace({' Not Available':' Emergency Room',' Not Mapped':' Emergency Room'})
```

```python
#replacing the null values in the coluumns with the mode from the columns
df['admission_type_id'] =  df['admission_type_id'].fillna('Emergency')
df['discharge_disposition_id'] =  df['discharge_disposition_id'].fillna('Discharged to home') 
df['admission_source_id'] =  df['admission_source_id'].fillna(' Emergency Room')
```

```python
#identifying row where all columns are missing diagnoses
df.loc[(df['diag_1']=='?') & (df['diag_2']=='?') & (df['diag_2']=='?'),['diag_1','diag_2','diag_3','number_diagnoses']]
```

```python
#replacing the columns that are just missing a diag_1 value with the mode
df.loc[(df['diag_1']=='?'),'diag_1'] = 428
```

```python
#identifying the columns where there is only 1 diagnosis and it is listed
df.loc[(df['diag_2']=='?') & (df['diag_3']=='?') & (df['number_diagnoses'] == 1) ,['diag_1','diag_2','diag_3','number_diagnoses']]
```

```python
#for the entries where there is only one diagnosis, replacing diag_2 and diag_3 with None
df.loc[(df['diag_2']=='?') & (df['diag_3']=='?') & (df['number_diagnoses'] == 1) ,['diag_2','diag_3']] = 'None'
```

```python
#dropping the rows who only have one diagnosis listed in diag_1 but number_diagnoses > 1
df.drop(df.loc[(df['diag_2']=='?') & (df['diag_3']=='?')].index, inplace = True)
```

```python
#figuring out the mode for the column
df['diag_2'].value_counts()
```

```python
#replacing the columns that are just missing a diag_2 value with the mode
df.loc[(df['diag_2']=='?'),'diag_2'] = 276
```

```python
#for the entries where there is two diagnosis and number_diagnoses=2, replacing diag_3 with None
df.loc[(df['diag_3']=='?') & (df['number_diagnoses'] == 2) ,['diag_2','diag_3']] = 'None'
```

```python
#dropping the rows who only have diagnosis listed in diag_1 and diag_2 but number_diagnoses > 3
df.drop(df.loc[(df['diag_3']=='?')].index, inplace = True)
```
